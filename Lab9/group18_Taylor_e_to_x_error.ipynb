## Taylor series of $e^x$

$$ e^{ x} = \sum_{n=0}^{\infty} \frac{x^n}{n!} $$

What is value of the partial Taylor series of $e^x$ up to the  $n$-th term of $e^x$ about $0$ (Maclaurin  series) for $x=2$

$$ e^{ x} \approx \sum_{i=1}^{n} \frac{x^i}{i!}  = 1 + x + \frac{x^2}{2!}+\cdots+ \frac{x^n}{n!}$$

```python
# ∑_{i=1}^n  x^i/i
import math

n = 10
x = 2
e_to_x = 0
for i in range(n+1):
    e_to_x += x**i/math.factorial(i) 
    print('e_to_x ≈ ∑_{i=1}^'+str(i),'x^i/i! =',e_to_x, )  
print('math.exp(x) = ',math.exp(x))
```

What is the minimun $n$ for which the $n+1$-th term of the Taylor series for $e^x$ about $0$ (Maclaurin  series) for $x$, is in absolute value less than  $\epsilon$.

$$\min_{n \in \mathbb{N}}\left|\frac{x^{n+1}}{(n+1)!}\right| < \epsilon$$

```python
# Retunrs Min n, |x^{n+1}/(n+1)!| < epsilon and ∑_{i=1}^{n+1}  x^i/i!|
import math

x = 1
epsilon = 10**(-5)
delta = 1
e_to_x = 0
i = 0
print("n".rjust(10)," ","∑_{i=1}^n  x^i/i!".center(21)," ","|x^n/n!|".center(20)," ","epsilon".center(20))
while  epsilon <= abs(delta):
  delta = x**i/math.factorial(i) 
  e_to_x += delta
  print(format(i, '10'),"  ", format(e_to_x, '.17f'),"  ",format( abs(delta), '.17f'),"  ",format(epsilon, '.17f')) 
  i = i+1
print("n = ",i-1) 
print('x = ',x)
print('e_to_x ≈ ∑_{i=1}^'+str(i-1),'x^i/i! =',e_to_x)    
print('math.exp(x) = ',math.exp(x))
```

## Taylor series of $ln(x+1)$

$$ ln(x+1) = \sum_{n=0}^{\infty} (-1)^{n-1}\frac{x^{n}}{n} $$

What is value of the partial Taylor series of $ln(x+1)$ up to the  $n$-th term of $ln(x+1)$ about $0$ (Maclaurin  series) for $x=0.4$

$$ ln(x+1) \approx \sum_{i=1}^{n} (-1)^{i-1}\frac{x^{i}}{i}  = x - \frac{x^2}{2} +\cdots+ (-1)^{n-1}\frac{x^{n}}{n}$$

```python
# ∑_{i=1}^n  [(-1)^(i-1)][(x^i/i)]
import math

n = 10
x = 0.4
ln_of_x_1 = 0.0
for i in range(1, n+1):
    ln_of_x_1 += ((-1)**(i-1))*(x**i/(i))
    print('ln_of_x+1 ≈ ∑_{i=1}^'+str(i),'[(-1)^('+str(i)+'-1)][(x^'+str(i)+'/'+str(i)+')] =',ln_of_x_1, )  
print('math.log(x+1) = ',math.log(x+1))
```

What is the minimun $n$ for which the $n+1$-th term of the Taylor series for $ln(x+1)$ about $0$ (Maclaurin  series) for $x$, is in absolute value less than  $\epsilon$.

$$\min_{n \in \mathbb{N}}\left|\frac{x^{n+1}}{(n+1)}\right| < \epsilon$$

```python
# Retunrs Min n, |x^{n+1}/(n+1)| < epsilon and ∑_{i=1}^{n+1}  [(-1)^(i-1)][(x^i/i)]
import math

x = 1
epsilon = 10**(-5)
delta = 1
ln_of_x_1 = 0
i = 1
print("n".rjust(10)," ","∑_{i=1}^n  [(-1)^(i-1)][(x^i/i)]".center(21)," ","|x^n/n|".center(20)," ","epsilon".center(20))
while  epsilon <= abs(delta):
  delta = ((-1)**(i-1))*(x**i/(i))
  ln_of_x_1 += delta
  if i < 100 or i > 99900:
    print(format(i, '10'),"  ", format(ln_of_x_1, '.17f'),"  ",format( abs(delta), '.17f'),"  ",format(epsilon, '.17f')) 
  i = i+1
print("n = ",i-1) 
print('x = ',x)
print('ln(x+1) ≈ ∑_{i=1}^'+str(i-1),'[(-1)^(i-1)][(x^i/i)] =',ln_of_x_1)    
print('math.log(x+1) = ',math.log(x+1))
```

# Error using the Lagrange's formula of the Residual $E_n$

For the Taylor series for $f(x) = ln(x+1)$ about $c=0$ (Maclaurin) 

the derivative $n+1$ of $ln(x+1) $ is $f^{(n+1)}(x) = (-1)^n(n!)(1+x)^{-(n+1)}$, so the error we used is:

$$ E_n = \frac{f^{(n+1)}(\xi)}{(n+1)!}(x-c)^{(n+1)} \le \frac{(-1)^n(n!)(1+x)^{-(n+1)}}{(n+1)!}x^{(n+1)} $$


we have to use de absolute value:

$$ \left|\frac{x^{n+1}}{(n+1)(x+1)^{n+1}}\right| $$


What is the minimun $n$ for which the error using the Lagrange's formula of the Residual $E_n$ of the Taylos series for $ln(x+1)$ about $0$ (Maclaurin  series), is less than  $\epsilon$.
 

```python
# Retunrs Min n, |x^{n+1}/(n+1)| < epsilon and ∑_{i=1}^{n+1}  [(-1)^(i-1)][(x^i/i)]|
import math

x = 0.9
epsilon = 10**(-12)
ln_of_x_1 = 0
i = 1
error = x**(i+1) / ((i+1)*(x+1)**(i+1))
print("n".rjust(10)," ","∑_{i=1}^n  x^i/i!".center(21)," ","error=ln(x+1) x^(n+1)/(n+1)!".center(20)," ","epsilon".center(20))
print(format(i, '10'),"  ", format(ln_of_x_1, '.17f'),"  ",format(error, '.17f'),"  ",format(epsilon, '.17f'))
while  epsilon <= abs(error):
  ln_of_x_1 +=  ((-1)**(i-1))*(x**i/(i))
  error = x**(i+1) / ((i+1)*(x+1)**(i+1))
  
  print(format(i, '10'),"  ", format(ln_of_x_1, '.17f'),"  ",format(error, '.17f'),"  ",format(epsilon, '.17f')) 
  i = i+1
print("n = ",i-1) 
print('x = ',x)
print('ln_of_x_1 ≈ ∑_{i=1}^'+str(i-1),'[(-1)^(i-1)][(x^i/i)] =',ln_of_x_1)    
print('math.log(x+1) = ',math.log(x+1))
```

## Taylor series of $sin(x)$

$$ sin(x) = \sum_{n=0}^{\infty} (-1)^{n}\frac{x^{2n+1}}{(2n+1)!} $$

What is value of the partial Taylor series of $sin(x)$ up to the  $n$-th term of $sin(x)$ about $0$ (Maclaurin  series) for $x=2$

$$ sin(x) \approx \sum_{i=0}^{n} (-1)^{i}\frac{x^{2i+1}}{(2i+1)!}  = x - \frac{x^3}{3!}+ \frac{x^5}{5!} +\cdots+ (-1)^{n}\frac{x^{2n+1}}{(2n+1)!}$$

```python
# ∑_{i=1}^n  [(-1)^n][(x^(2i+1))/((2i+1)!)]
import math

n = 10
x = 2
sin_of_x = 0
for i in range(n+1):
    sin_of_x += ((-1)**i)*x**(2*i+1)/math.factorial(2*i+1) 
    print('sin_of_x ≈ ∑_{i=1}^'+str(i),'[(-1)^n][(x^(2i+1))/((2i+1)!)] =',sin_of_x, )  
print('math.sin(x) = ',math.sin(x))
```
